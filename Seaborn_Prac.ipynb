```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#To get names of in built seaborn dataset
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
data.head(5)
```

```python
#Axes level approach
#Scatterplot
sns.scatterplot(data=data,x='tip',y='total_bill')
```

```python
#Function level approach
sns.relplot(data=data,x='tip',y='total_bill',kind='scatter',hue='sex',style = 'smoker',size='size')
```

```python
#Line Plot
sns.lineplot(data=data,x='tip',y='size')
```

```python
#Histogram
sns.histplot(data=data,x='tip',bins=20)
```

```python
#KDE Plot
sns.kdeplot(data=data,x='tip')
```

```python
#ECDF Plot
sns.ecdfplot(data=data,x='tip')
```

```python
#Barplot
#Axes approach
sns.barplot(data=data,x='size',y='total_bill')
```

```python
#Function approach
sns.catplot(data=data,x='size',y='total_bill',kind='bar',hue='sex')
```

```python
#Count
sns.catplot(data=data,x='smoker',kind='count',hue='sex')
```

```python
#Boxplot
sns.boxplot(data=data,x='size',y='total_bill')
```

```python
#Violinplot
sns.violinplot(data=data,x='size',y='total_bill')
```

```python
#Swarmplot
sns.swarmplot(data=data,x='size',y='tip')
```

```python
corr = data.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
g = sns.FacetGrid(data,col = 'sex',row = 'time')
g.map(sns.scatterplot,'total_bill','tip')
```

```python
#Catplot multigraph
sns.catplot(data=data,x='day',y='total_bill',kind='box',col='sex')
```

```python
#Pairplot
sns.pairplot(data,hue='sex',height=2,aspect=1)
```
